```python
import sys
import os
import glob
import re 
import pandas as pd
import numpy as np
from lxml import etree
import matplotlib.pyplot as plt
import requests
import time
from collections import Counter
```

```python
from transformers import BertTokenizer, AutoTokenizer
```

```python
#tz_ = BertTokenizer.from_pretrained("dbmdz/bert-base-multilingual-cased-finetuned-conll03-spanish")
bert_multilingual_cased_tz = BertTokenizer.from_pretrained("bert-base-multilingual-cased")
bert_cased_tz = BertTokenizer.from_pretrained("bert-base-cased")
mt5_base = AutoTokenizer.from_pretrained("google/mt5-base")
xlm_roberta_base = AutoTokenizer.from_pretrained("xlm-roberta-base")
```

```python
if not os.path.exists("./../data/translations_tokenized/"):
    os.makedirs("./../data/translations_tokenized/")
```

```python
errors = []

for bible_path in glob.glob("./../data/translations_joined_corpus/*.tsv"):

    bible_name  = os.path.splitext(os.path.split(bible_path)[1])[0]
    print(bible_name)

    bible = pd.read_csv(bible_path, sep="\t").fillna("")

    try:
        for index, row in bible.iterrows():
            bible.loc[index, "bert_multilingual_cased"] = "|".join(bert_multilingual_cased_tz.tokenize(row["text"]))
            bible.loc[index, "bert_cased"] = "|".join(bert_cased_tz.tokenize(row["text"]))
            bible.loc[index, "mt5_base"] = "|".join(mt5_base.tokenize(row["text"]))
            bible.loc[index, "xlm_roberta_base"] = "|".join(xlm_roberta_base.tokenize(row["text"]))

        print(bible.sample(5, random_state=2021))


        bible.to_csv("./../data/translations_tokenized/" + bible_name + ".tsv", sep = "\t" )
    except:
        errors += bible_name

errors
```
