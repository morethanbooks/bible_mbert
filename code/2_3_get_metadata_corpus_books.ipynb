```python
errors = []

i = 0

original_length = len(metadata)


metadata_books = metadata.head(0).copy()

for index, row in metadata.iterrows():
    try:
        print(row["file_name"])

        df = pd.read_csv("./../data/translations_tokenized_counts/" +row["file_name"] + ".tsv", sep="\t", index_col=0)

        for book in df["codebook"].unique().tolist():

            book_file_name = str(book) + "_" + row["file_name"] 
            book_file_name = re.sub(r"\.", r"", book_file_name)

            metadata_books.loc[i ] = row
            metadata_books.loc[i, "book"] = str(book)
            metadata_books.loc[i, "book_file_name"] = book_file_name

            book_df = df.loc[df["codebook"] == str(book)].copy()

            metadata_books.loc[i, "text_first_verse"] = book_df.head(1)["text"].values.tolist()[0]
            metadata_books.loc[i, "id_first_verse"] = book_df.head(1)["id"].values.tolist()[0]


            for column in columns:
                metadata_books.loc[i, column + "_mean"] = book_df[column].mean()
                metadata_books.loc[i, column + "_median"] = book_df[column].median()
                metadata_books.loc[i, column + "_std"] = book_df[column].std()

            #metadata.loc[original_length  + i, column + "_mean / text_count_tokens_mean"] = metadata[column + "_mean" ] / metadata["text_count_tokens_mean" ]

            i += 1
            book_df.to_csv("./../data/books_tokenized/" + book_file_name + ".tsv", sep="\t" )

    except:
        errors.append(row["file_name"])
```
